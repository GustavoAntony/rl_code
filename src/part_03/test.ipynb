```python
import matplotlib.pyplot as plt
import numpy as np

# Cria dados de exemplo
data1 = np.random.normal(0, 1, 1000)
print(data1)
data2 = np.random.normal(2, 1, 1000)
data3 = np.random.normal(4, 1, 1000)

# Plotar histograma com cores diferentes
plt.hist([data1, data2, data3], bins=20, density=True, color=['red', 'blue', 'green'], alpha=0.5, label=['Data 1', 'Data 2', 'Data 3'])

# Configurações do gráfico
plt.legend()
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Histograma com cores diferentes')

# Mostrar o gráfico
plt.show()
```
